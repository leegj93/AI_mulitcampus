```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
correlation = df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cmap='cubehelix', annot=True) # cmap: 색상 설정, annot: annotation 옵션은 수치를 보여줄지 말지
```

```python
# 주요한 변수는 left인데, 이 것이 회사를 관둘지 말지가 매우 중요함
```

```python
df2 = pd.get_dummies(df['sales'])
df3 = pd.get_dummies(df['salary'])
```

```python
df_new = df.copy()
```

```python
df_new[df2.columns]=df2
df_new[df3.columns]=df3
```

```python
correlation = df_new.corr()
```

```python
correlation['left'].sort_values()
```

```python
correlation.loc['left',df2.columns].sort_values()
```

```python
correlation.loc['left',df3.columns].sort_values()
```
