```python
from sklearn.datasets import load_digits
digits=load_digits()
samples=[0,10,20,30,1,11,21,31]#이미지번호
```

```python
digits
```

```python
d=[]
for i in range(8):
    d.append(digits.images[samples[i]])#[0,10,20,30,1,11,21,31]
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(d[i], cmap=plt.cm.bone_r, interpolation='nearest')
    plt.title("img {}".format(i+1))
    plt.grid(False); plt.xticks([]);plt.yticks([]);
plt.show()
```

```python
#2차원 이미지(8,8) -> 1차원 벡터(크기 64,1)
#d[0].shape
v=[]
for i in range(8):
    v.append(d[i].reshape(64,1)) #벡터화
    #print("-"*50)
#np.shape(v) #(8, 64, 1)
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(v[i],aspect=0.2, cmap=plt.cm.bone_r, interpolation='nearest')
    plt.title("img {}".format(i+1))
    plt.grid(False); plt.xticks([]);plt.yticks([]);
plt.tight_layout(w_pad=3)
plt.show()

#1. 연습문제
#벡터의 내적을 수행 -> 가장 큰 내적을 출력(이미지 번호)
```

```python
from scipy import misc
img_rgb=misc.face()
img_rgb.shape #(가로,세로,채널)

plt.subplot(221)
plt.imshow(img_rgb[ : , : ,0], cmap=plt.cm.gray)
plt.axis("off")
plt.title("red img")

plt.subplot(222)
plt.imshow(img_rgb[ : , : ,1], cmap=plt.cm.gray)
plt.axis("off")
plt.title("green img")

plt.subplot(223)
plt.imshow(img_rgb[ : , : ,2], cmap=plt.cm.gray)
plt.axis("off")
plt.title("blue img")

plt.subplot(224)
plt.imshow(img_rgb, cmap=plt.cm.gray)
plt.axis("off")
plt.title("rgb img")

plt.show()
```
